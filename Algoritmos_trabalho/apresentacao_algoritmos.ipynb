<h1>Plotagem dos algoritmos </h1>

Importação do módulo para construir os gráficos

```python
import matplotlib.pyplot as plt
```

<h3>Gráfico dos Algoritmos Simples</h3>

```python
sizes = [5000, 10000, 50000, 100000, 500000]

bubble_sort_times = [3.3509, 15.1669, 343.4534, 1343.6017, 8755.0000]
insertion_sort_times = [1.2760, 7.0000, 149.8476, 738.4283, 9480.0000]
selection_sort_times = [1.9600, 6.9620, 256.3533, 1576.8519, 7200.0000]

# Configurar os gráficos
plt.figure(figsize=(10, 6))

# Calcular a posição das barras no eixo x
x = range(len(sizes))

# Plotar as barras para o Bubble Sort
plt.bar(x, bubble_sort_times, label='Bubble Sort', color='red', width=0.2)

# Plotar as barras para o Insertion Sort
plt.bar([i + 0.2 for i in x], insertion_sort_times, label='Insertion Sort', color='blue', width=0.2)

# Plotar as barras para o Selection Sort
plt.bar([i + 0.4 for i in x], selection_sort_times, label='Selection Sort', color='green', width=0.2)

# Definir os rótulos dos eixos
plt.xlabel('Tamanho da Entrada (N)')
plt.ylabel('Tempo de Execução (s)')

# Definir os rótulos do eixo x
plt.xticks([i + 0.2 for i in x], sizes)

# Adicionar uma legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
